```python
import pandas as pd
```

```python
master_items = pd.read_json('MasterItems_split.json') #split removd the first "MasterItems" = { } so the content addresses are the columns by default now
display(master_items.iloc[:5])
print(master_items.columns)
```

```python
# match all if it starts with
# m: melee
# r: ranged
master_items_attachments_only = master_items.filter(regex='content/items/weapons/player/[mr].*')
# print first 5 columns as test
display(master_items_attachments_only.iloc[:, 0:5])

display(master_items_attachments_only.loc['attach_node'])
```

```python
# creates a series with only address and attach_nodes
#   Series
MI_attach_nodes = master_items_attachments_only.loc['attach_node']
# filter to only unique entries
#   Array
MI_attach_nodes = MI_attach_nodes.unique()
MI_attach_nodes = sorted(MI_attach_nodes)

# Using open() function
file_path = 'weapon_attachments_list_of_attach_nodes.txt'

# Open the file in write mode
with open(file_path, 'w') as file:
    for node in MI_attach_nodes:
        # strip the string identifier part
        node = node.replace(' (string)', '')
        file.write(node+'\n')
        print(node)
```
